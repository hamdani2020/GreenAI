Training Model

```python
!pip install ultralytics
import cv2
from ultralytics import YOLO

# small https://github.com/ultralytics/assets/releases/download/v8.3.0/yolo11m.pt

_project = "runs/CropDisease"
_name = "Baseline"
_data = "/content/data.yaml"

# Epoch
_epochs = 30
model = YOLO('yolo11m.pt')

results = model.train(
    data = _data,
    epochs = _epochs,
    batch = 16,
    project = _project,
    name = _name
)
```

**Predicting**

```python
model_pt = "/content/runs/CropDisease/Baseline/weights/best.pt"
model = YOLO(model_pt)

results = model.predict(
    source="/content/train/images",
    conf=0.5,
    save=True,
)
```
